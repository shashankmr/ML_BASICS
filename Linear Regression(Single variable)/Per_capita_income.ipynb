```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model as lm
```

```python
df =pd.read_csv("income.csv")
```

```python
%matplotlib inline
plt.xlabel("year", fontsize = 25)
plt.ylabel("income(US$)",fontsize = 25)
plt.scatter(df.year,df.income, color = "red", marker = "x")
```

```python
lr = lm.LinearRegression()
lr.fit(df[["year"]],df.income)
```

```python
lr.score(df[["year"]],df.income)
```

```python
lr.predict(2050)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
%matplotlib inline
plt.xlabel("year", fontsize = 25)
plt.ylabel("income(US$)",fontsize = 25)
plt.scatter(df.year,df.income, color = "red", marker = "x")
plt.plot(df.year,lr.predict(df[["income"]]))
```
